```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = [1, 2, 3, 4, 5]
y = [0,2, 4, 6, 8]

# label is apart of the legend
# color can be name and hex code
#plt.plot(x, y, label='2x', color='green', linewidth= 4, marker='.',markersize=10, linestyle='--', markeredgecolor='blue')

# Short hand notation of above
# fmt = '[color][marker][line]'
plt.plot(x,y, 'r.--', label='2x')

## Line Number Two
x2 = np.arange(0,4,0.5)
plt.plot(x2, x2**2, 'b', label='x^2')


plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# if you want to change ticks
plt.xticks([0, 1, 2, 3, 4])
#plt.yticks([0, 2, 4, 6, 8, 10])

# calling the legend
plt.legend()
plt.show()
```
