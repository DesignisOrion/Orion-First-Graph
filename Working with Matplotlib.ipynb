```python
# Import things
import matplotlib.pyplot as plt

# allows you to see the plots in the notebook
%matplotlib inline

# some data
x = list(range(0, 10))
y = list(range(-10, 0))

plt.plot(x, y)

```

```python
a = [0, -100, 25, 67, -323]
b = [0 ,3, 7, 3, 9]
plt.plot(a, b)
```

```python
# if you want to only focus on the triangle
plt.axis([-50, 80, 2, 8])
# if you want to change the x-axis ticks at the bottom
plt.xticks((-40, -20, 0, 20, 40, 60, 80), ('h','e', 'l', 'l', 'o'))
#plt.plot(a, b)

# To change color use "color = red" inside the plt.plot()

plt.plot(a, b, color="red")
```

```python
# placing title of the graph
plt.title("Triangle")
plt.xlabel("Array A")
plt.ylabel("Array B")
plt.plot(a, b)
```
